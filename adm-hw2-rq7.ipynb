```python
#The Pareto principle states that for many outcomes roughly 80% of consequences come from 20% of the causes. 
#Also known as 80/20 rule, in e-commerce simply means that most of your business, around 80%, likely comes from about 20% of your customers.
#Prove that the pareto principle applies to your store.

df.reset_index(inplace = True)
import math
Descending_purchase_by_user = df[(df.event_type == 'purchase')].groupby([df.user_id]).agg(Price = ('price', 'sum')).sort_values('Price',ascending= False)
Twenty_Percent = Descending_purchase_by_user.iloc[:int(len(Descending_purchase_by_user)*0.2)+1, :].Price.sum()
Twenty_Percent
```

```python
Total = df[(df.event_type == 'purchase')].price.sum()
Total
print('20% of customers consequences has the ' + str((Twenty_Percent/Total*100)) + '% percentage of total.')
```
